### start

- 爬虫流程
1. 获取网页内容
2. 解析网页内容
3. 储存和分析数据

- 两个禁区
1. 不能干扰网页的的正常运行
2. 不能爬收到法律禁止的数据和信息

- 接下来要学的
1. 正则表达式
2. 多线程

- 爬虫类型
1. 通用
2. 聚焦
3. 增量式

### robots.txt协议
- 君子协议
- 规定了哪些数据可以爬取，哪些不能被爬取
- 例如淘宝的https://world.taobao.com/robots.txt

### http协议
- 请求头信息
  1. User-Agent
  2. Connection

- 响应头信息
  1. Content_Typt

- https
  1. 安全的超文本传输协议
  2. 数据交互是有数据加密

- 加密方式
  1. 对称密钥加密
  2. 非对称密钥加密
  3. 证书密钥加密

### requests模块
- python中原生的网络请求模块，功能强大，效率高
- 作用：模拟浏览器发送请求
- 如何使用
  1. 指定url
  2. 发起请求
  3. 获取响应数据
  4. 持久化存储数据

### 数据解析
- 分类
  1. 正则
  2. bs4
  3. `xpath`
- 原理
  解析的局部文本内容都会在标签之间或者标签对应的属性中存储
  1. 标签定位
  2. 标签或标签对应的属性中存储的数据值进行提取（解析）
